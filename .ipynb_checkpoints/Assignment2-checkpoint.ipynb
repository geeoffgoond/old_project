```python
#%% specifying data location data from gitHub, preparing data

import os
folderData="https://raw.githubusercontent.com/geeoffgoond/data/master/"
fileCSV=os.path.join(folderData,"thads2013m.csv")
#fileCSV="https://raw.githubusercontent.com/geeoffgoond/data/master/thads2013m.csv"

# importing data into pandas
import pandas as pd
dataCSV=pd.read_csv(fileCSV)

#setting dataset name, checking datatypes
hads=dataCSV
hads.dtypes
```

```python
#%%variable of interest: housing burden, which is housing cost as a fraction of income
hads["BURDEN"].dtypes
```

```python
hads["BURDEN"].value_counts() #checking data
#we can see that we have missing data, coded as -9
#we also have data for households with no income coded as -1
```

```python
#recoding missing values:
import numpy as np
NA1=[-9.000000]
hads["BURDEN"].replace(NA1,np.nan,inplace=True)
```

```python
#getting descriptives
hads["BURDEN"].describe()
```

```python
# checking standard deviation, skewness, kurtosis
hads["BURDEN"].std()
```

```python
hads["BURDEN"].skew()
```

```python
hads["BURDEN"].kurt
```

```python
#plotting data
import seaborn as sns
from scipy.stats import norm
import matplotlib.pyplot as plt # needed for more customization

sns.distplot(hads["BURDEN"].dropna(),kde=False, fit=norm)
plt.legend(('data fit to normal','data as it is'))
plt.title("Housing Burden Plot")

plt.show()
```

```python
#plotting with mean and median:
mnVar=hads["BURDEN"].mean()
mdVar=hads["BURDEN"].median()

sns.distplot(hads["BURDEN"].dropna(),kde=False,fit=norm)
plt.title('Housing Burden Plot')
plt.axvline(mnVar, color='b', linestyle='dashed', linewidth=2,label='mean')
plt.axvline(mdVar, color='r', linestyle='dashed', linewidth=2,label='median')
plt.legend()
```

```python
#boxplot
plot,dataBP=hads["BURDEN"].plot.box(vert=False,return_type='both')  
```

```python
#just for exploratory purposes, I want to see what it looks like when values are restricted to between 0 and 1, or no rent burden vs 100% burden
NA2=[-1.000000]
hads["BURDEN"].replace(NA2,np.nan,inplace=True)

#getting rid of values > 1 since housing burden should not be able to exceed income
for value in hads["BURDEN"]:
    if value > 1 or value <0:
        NAv=[value]
        hads["BURDEN"].replace(NAv,np.nan,inplace=True)
```
